# loading the dataframe

```python
df = pd.read_csv("wines.csv")
```

```python
df
```

# counting how many of those had 100 ratings

```python
ratings = df.loc[df["points"] == 100].to_dict(orient="records")
```

```python
ratings
```

```python
count = 0
for rate in ratings:
    count = count + 1
print(count)
```

# finding out the maximum price

```python
max_price = df["price"].max()
```

```python
max_price_data = df.loc[df["price"] == max_price].to_dict(orient="records")
print(max_price_data)
```
